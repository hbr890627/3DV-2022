```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
# point cloud
# plot the loss curve
# with open("checkpoints/loss_point_1.pkl", "rb") as f:
with open("checkpoints/loss_point.pkl", "rb") as f:
    hist = pickle.load(f)

    plt.figure(figsize=(15, 5))
    plt.title("Point - Loss curve")
    plt.xlabel("step")
    plt.ylabel("loss")

    plt.plot(hist["loss"])
```

```python
# voxel grid
# plot the loss curve
# with open("checkpoints/loss_voxel_1.pkl", "rb") as f:
with open("checkpoints/loss_voxel.pkl", "rb") as f:
    hist = pickle.load(f)

    plt.figure(figsize=(15, 5))
    plt.title("Voxel - Loss curve")
    plt.xlabel("step")
    plt.ylabel("loss")

    plt.plot(hist["loss"])
```
